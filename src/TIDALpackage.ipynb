```python
import numpy as np
import sympy as sym


import _MEWtools as MEW


r, θ, φ, t = sym.symbols(  'r θ φ t'    ,   real = True  )
```

Still Need to include visco phase shift

```python
europa_4 = MEW.satellite(

    # The structure input 
    
    rin = [1.394e6, 1.546e6, 1.556e6, 1.561e6],   # <m>      a list of bounding radii 
    pin = [3824   , 1000   , 1000   , 1000 ],     # <kg/m^3> a list of density 
    uin = [65e9   , 0      , 3.5e9  , 3.5e9  ],   # <Pa>     a list of rigidity    (0 means fluid)
    nin = [0      , 0      , 1e13      , 0],         # <Pa s>   a list of viscosity   (0 means not viscoelastic)

    

    # The orbit input

    G  = 6.67e-11,             # <N m^2/kg^2> a constant of nature
    M  = 1.898e27,             # <kg>         source mass
    a  = 6.711e8,              # <m>          semimajor distance

    e  = 0.01,                 #              Orbital Eccentricity
 
    
    # The spin input    
    
    o  = 0,                    # <rad>        Obliquity
    op = 0,                    # <rad>        Phase of Obliquity (0° - spin pole points toward Jupiter at peri)
    on = 0,                    # <1/s>        Obliquity phase rate (spin pole precession)

    Ω  = 0,                    # <1/s>        spin rate (set to 0 for synchronous, 85 hrs for Europa) 

    NSRn  = 0,                 # <1/s>        non synchronous rotation rate (2 pi / N yrs)

    libA  = 0,                 # <rad>        libration Amplitude 
    libp  = 0,                 # <rad>        libration phase
    libn  = 0                  # <1/s>        libration frequency 
    
                     )
```

tt - theta theta <br>
pp - phi phi     <br>
tp - theta phi   <br>

```python
display(europa_4.tt.subs(t,0).subs(φ,np.pi/4).subs(θ,np.pi/4).expand())
display(europa_4.pp.subs(t,0).subs(φ,np.pi/4).subs(θ,np.pi/4).expand())
display(europa_4.tp.subs(t,0).subs(φ,np.pi/4).subs(θ,np.pi/4).expand())
```

PC1 - PC1    <br>
PC2 - PC2    <br>

PCΨ  - PC1 orientation  tan[y/x] <br>
PCΨ2 - PC1 orientation  tan[y,x] <br>

```python
display(europa_4.PC1.subs(t,0).subs(φ,6*np.pi/12).subs(θ,4*np.pi/12).expand())
display(europa_4.PC2.subs(t,0).subs(φ,6*np.pi/12).subs(θ,4*np.pi/12).expand())

display(sym.N(europa_4.PCΨ.subs(t,0).subs(φ,6*np.pi/12).subs(θ,4*np.pi/12).expand())*(180/np.pi))
display(sym.N(europa_4.PCΨ2.subs(t,0).subs(φ,6*np.pi/12).subs(θ,4*np.pi/12).expand())*(180/np.pi))

# angle is for PC1, 0° is north, varies from -90 (east) to 90 (west)
```
